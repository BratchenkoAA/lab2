```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_02_13.
#
# Выполнил: Братченко А.А.
# Группа: ЦИБ-241

import math


a = int(input('a = '))
b = int(input('b = '))

n_sum = sum(range(a, b +1 ))
n_mult= math.prod(range(a, b+1))
n_avg= n_sum/ (b-a+1)

n_avg_geom= n_mult**(1/2)

odd_numbers = [x for x in range(a, b + 1) if x % 2 != 0]

if odd_numbers:
    n_avg_geom = math.prod(odd_numbers) ** (1 / len(odd_numbers))



print("Сумма =", n_sum)
print("Произведение =", n_mult)
print("Среднее арифметическое =", round(n_avg, 2))
print("Среднее геометрическое нечетных чисел =", round(n_avg_geom, 2))

# --------------
# Пример вывода:
#
# a = 1
# b = 5
# Сумма = 15
# Произведение = 120
# Среднее арифметическое = 3.00
# Среднее геометрическое нечетных чисел = 2.47
```

```python
# Задание task_02_29.
#
# Выполнил: Братченко А.А.
# Группа: ЦИБ-241

import math


n = int(input("n = "))

nums=[]

for i in range(n):
  num = float(input(f'Введите {i+ 1}-й элемент списка: '))
  nums.append(num)

print(f'Исходный список: {nums}')


nums_neg= []

nums_pos = [x for x in nums if x > 0]
print(f'Положительные числа: {nums_pos}')

nums_neg= [x for x in nums if x < 0]
print(f' Отрицательные числа: {nums_neg}')

if nums_pos:
    sr_ar = sum(nums_pos) / len(nums_pos)
    print(f'Ср. арифм.: {sr_ar:.2f}')
else:
    print('Нет положительных чисел для расчета среднего арифметического.')

if nums_neg:
    sr_geom = math.prod(nums_neg) ** (1 / len(nums_neg))
    print(f'Ср. геом.: {sr_geom:.2f}')

# --------------
# Пример вывода:
#
# n = 4
# Введите 1-й элемент списка: 1
# Введите 2-й элемент списка: 6
# Введите 3-й элемент списка: -10
# Введите 4-й элемент списка: -3
# Исходный список: [1.0, 6.0, -10.0, -3.0]
# Положительные числа: [1.0, 6.0]
# Отрицательные числа: [-10.0, -3.0]
# Ср. арифм.: 3.50
# Ср. геом.: 5.48
```

```python
# Задание task_02_30.
#
# Выполнил: Братченко А.А.
# Группа: ЦИБ-241



# Все разделенные пробелом элементы будут преобразованы в список целых чисел
nums = [int(item) for item in input().split()]

# 1. Все положительные
for item in nums:
    if item <= 0:
        all_pos_1 = False
        break
else:
    all_pos_1 = True

all_pos_2 = all([item > 0 for item in nums])

# 2. Хотя бы 1 нулевой элемент
for item in nums:
    if item == 0:
        any_zero_1 = True
        break
else:
    any_zero_1 = False

any_zero_2 = any([item == 0 for item in nums])

# 3. Все четные
# Удалите комментарий и допишите код (определите переменные all_even_1, all_even_2)
for item in nums:
    if item %2 != 0:
        all_even_1 = False
        break
else:
   all_even_1 = True
all_even_2 = all(item % 2 == 0 for item in nums)

# 4. Хотя бы 1 нечетный элемент
# Удалите комментарий и допишите код (any_odd_1, any_odd_2)

for item in nums:
    if item % 2 != 0:
        any_odd_1 = True
        break
else:
    any_odd_1 = False

any_odd_2 = any(item % 2 != 0 for item in nums)

print("Все положительные:", all_pos_1, all_pos_2)
print("Хотя бы 1 нулевой элемент:", any_zero_1, any_zero_2)
print("Все четные:", all_even_1, all_even_2)
print("Хотя бы 1 нечетный элемент:", any_odd_1, any_odd_2)

# --------------
# Пример вывода:
#
# -1 1 100 0
# Все положительные: False False
# Хотя бы 1 нулевой элемент: True True
# Все четные: False False
# Хотя бы 1 нечетный элемент: True True
```

```python
# Задание task_02_34.
#
# Выполнил: Братченко А.А.
# Группа: ЦИБ-241



# 1. Заполнение списка

n = int(input('Количество банков = '))

deposits = []

# 2. Самый доступный банк (с наименьшей первоначальной суммой)

# Сбор данных о вкладах
for _ in range(n):
    entry = input('Введите данные о вкладе (Банк Сумма Процент): ')
    name, initial_sum, rate = entry.split()
    initial_sum = int(initial_sum)  # Преобразуем сумму в целое число
    rate = float(rate)  # Преобразуем процент в вещественное число

    # Добавляем данные в список как словарь
    deposits.append({
        "name": name,
        "initial_sum": initial_sum,
        "rate": rate
    })

# 2. Самый доступный банк (с наименьшей первоначальной суммой)
most_accessible_bank = min(deposits, key=lambda x: x['initial_sum'])

# 3. Самый выгодный банк (с наибольшей годовой прибылью)
most_profitable_bank = max(deposits, key=lambda x: x['initial_sum'] * x['rate'] / 100)

# Вывод результатов
print(f"Самый доступный банк: {most_accessible_bank['name']}, первоначальная сумма: {most_accessible_bank['initial_sum']:.2f} руб.")
print(f"Самый выгодный банк: {most_profitable_bank['name']}, прибыль: {(most_profitable_bank['initial_sum'] * most_profitable_bank['rate'] / 100):.2f} руб.")


# 3. Самый выгодный банк
#    прибыль = сумма * процент / 100

# Удалите комментарий и допишите код

# --------------
# Пример вывода:
#
# Количество банков = 3
# Банк_1 50000 5.2
# Банк_1 50000 5.2
# Банк_3 45000 6.8
# Самый доступный банк: Банк_2, первоначальная сумма: 20000.00 руб.
# Самый выгодный банк: Банк_3, прибыль: 3060.00 руб.
```
